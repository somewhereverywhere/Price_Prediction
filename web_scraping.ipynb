```python
from bs4 import BeautifulSoup
import requests

url = "https://www.yelp.com/menu/village-the-soul-of-india-hicksville"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# 200 indicates succes
print(soup.prettify())
print(response.status_code)
```

```python

# Find all <a> tags that contain menu item links
menu_items = []

# Find all <a> tags that contain menu item links
menu_sections = soup.find_all('h4')

for section in menu_sections:
    # Look for an <a> tag inside the <h4> (for items that have links)
    link = section.find('a', href=True)

    if link:  # If there is an <a> tag, get the item name from it
        item_name = link.text.strip()
        # Check if it's a valid menu item and not related to photos or reviews
        if link['href'].startswith("/menu/village-the-soul-of-india-hicksville/item/") and "#menu-reviews" not in link['href'] and "#photos" not in link['href']:
            menu_items.append({"name": item_name, "price": None})  # Add the item name to the list
    else:
        # If there is no <a> tag, get the item name directly from the <h4> text
        item_name = section.text.strip()
        if item_name:
            menu_items.append({"name": item_name, "price": None})
# Find menu item prices (adjust class names as needed)
menu_sections = soup.find_all('div', class_="menu-item-prices arrange_unit")  # Replace with actual container class for prices

# Assign prices to the corresponding menu item
for i, section in enumerate(menu_sections):
    try:
        # Extract price for each menu item
        item_price = section.find('li', class_="menu-item-price-amount").text.strip()  # Replace with actual class
        if i < len(menu_items):
            menu_items[i]["price"] = item_price  # Update the price in the dictionary
    except AttributeError:
        continue  # Skip if price is not found

# Output the scraped menu
print("Scraped Menu:")
for item in menu_items:
    name = item.get("name", "N/A")
    price = item.get("price", "N/A")
    print(f"{name}: {price}")

```

```python
import csv

csv_filename = "village_.csv"

# Open the CSV file in write mode
with open(csv_filename, mode='w', newline='', encoding='utf-8') as file:
    # Define the CSV writer
    writer = csv.DictWriter(file, fieldnames=["name", "price"])

    # Write the header row (column names)
    writer.writeheader()

    # Write the menu items to the CSV file
    for item in menu_items:
        writer.writerow(item)

print(f"Menu items have been written to {csv_filename}")
```
